```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
path_to_lammps_tutorial = "/home/simon/Git/Lammps/lammpstutorials.github.io/docs/sphinx/source/tutorials/figures/level0/lennard-jones-fluid/"
```

```python
from utilities_plot import myblue, lightgray, darkgray, complete_panel
```

```python
total_energy = np.loadtxt('totalenergyvstime.dat')
kinetic_energy = np.loadtxt('kineticenergyvstime.dat')
potential_energy = np.loadtxt('potentialenergyvstime.dat')
```

```python
figure_name = 'exercice-energy'
for mode, mygray in zip(['light', 'dark'], [lightgray, darkgray]):

    fig = plt.figure(figsize=(16, 12))

    ax, n = [], 0
    l_tot, c_tot = 3, 1

    n += 1
    ax.append(plt.subplot(l_tot, c_tot, n))
    ax[-1].plot(total_energy.T[0]*0.005, total_energy.T[1], '-', color=myblue, linewidth=2)
    complete_panel(ax[-1], None, r'Tot. energy', legend=False, color=mygray)

    n += 1
    ax.append(plt.subplot(l_tot, c_tot, n))
    ax[-1].plot(kinetic_energy.T[0]*0.005, kinetic_energy.T[1], '-', color=myblue, linewidth=2)
    complete_panel(ax[-1], None, r'Kin. energy', legend=False, color=mygray)

    n += 1
    ax.append(plt.subplot(l_tot, c_tot, n))
    ax[-1].plot(potential_energy.T[0]*0.005, potential_energy.T[1], '-', color=myblue, linewidth=2)
    complete_panel(ax[-1], r'Time', r'Pot. energy', legend=False, color=mygray)

    fig.tight_layout()
    if mode == 'light':
        plt.savefig(path_to_lammps_tutorial+figure_name+'-light.png', bbox_inches = 'tight', pad_inches = 0.057, transparent=True)
        plt.savefig(figure_name+'-white.png', bbox_inches = 'tight', pad_inches = 0.057)
    else:
        plt.savefig(path_to_lammps_tutorial+figure_name+'-dark.png', bbox_inches = 'tight', pad_inches = 0.057, transparent=True)
    
    plt.show()
```
